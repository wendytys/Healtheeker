```python
import pandas as pd
```

```python
# f1 = "AUSNUT 2011-13 AHS Food Nutrient Database.xls"
f2 = "Release 1 - Food nutrient database.xlsx"

# file1 = pd.read_excel(f1)
file2 = pd.read_excel(f2)
```

```python
file2 = pd.ExcelFile(f2)
file2.sheet_names
```

```python
file21 = file2.parse("All solids & liquids per 100g")
file21.head()
```

```python
col = ['Food Name', 'Nitrogen', 'Total dietary fibre', 'Total Fat', 
       'Fructose', 'Glucose', 'Sucrose', 'Maltose', 'Lactose', 'Total sugars','Sorbitol'] 
#        'Added sugars', 'Free sugars']
file22 = file21[col].loc[1:,:]
file23 = file22.fillna(0)#.dropna()
file23.head()
```

```python
file23.shape
```

# 1.Fruit

```python
# fructose_sort[fructose_sort.index == 1290]
# fructose_sort.loc[[1290],:]
```

```python
fructose_sort = file23.sort_values(by=['Sorbitol','Fructose'], ascending=False).head(200)#.values

fructose_sort = fructose_sort[fructose_sort['Food Name'].str.contains('raw') | 
                              fructose_sort['Food Name'].str.contains('Raw') ]
# fructose_sort = fructose_sort.sort_values(by=['Sorbitol'], ascending=False).head(50)
# fructose_sort
fructose_sort.shape#['Food Name'].values
fructose_sort.head(100)
```

```python
import re
n = 0
f = []
fruit = []
for i in range(fructose_sort.shape[0]):
    name = fructose_sort['Food Name'].values[i]
#     print(name.split())
    name = name.split()[0]
    name = name.replace(',','')
    # name = name[0].upper() + name[1:].lower()
    
    if name not in f:
        f.append(name)
        
        f10 = fructose_sort[fructose_sort['Food Name'].str.contains(name)]
        info = [name] + f10.loc[:,f10.columns[1:]].mean().tolist()
        fruit.append(info)
    
```

```python
fruit_df = pd.DataFrame(fruit,columns=file23.columns).sort_values(by=['Total sugars'], ascending=False)
fruit_df = fruit_df.sort_values(by=['Sorbitol'], ascending=False).reset_index(drop=True)

fruit_df
```

```python
fruit_df = fruit_df.loc[[0,1,2,3,4,5,13,18,21,22,23,25,27,28,30,31,39],:].reset_index(drop=True).round(1)
fruit_df = fruit_df.sort_values(by=['Total sugars'], ascending=False)
fruit_df
```

```python
fruit_df = fruit_df.loc[[16,11,12,15,5,1,0,9,4,3],:].reset_index(drop=True)
fruit_df
```

```python
veges = "Garlic, garlic salt, garlic powder, Onions, onion powder, small pickled onions, Beetroot, \
Artichoke, Asparagus,Baked beans, Black eyed peas, Broad beans, Butter beans\
, Cassava, Cauliflower, Celery, Choko, Falafel, Fermented cabbage, sauerkraut\
,Haricot beans, Kidney beans, Lima beans, Leek bulb, Mange Tout, Mixed vegetables, Mung beans, \
Mushrooms, Peas, sugar snap, Pickled vegetables, Red kidney beans, Savoy Cabbage, \
Soy beans, soya beans, Split peas, Scallions, spring onions, Shallots, Taro"
veges = veges.replace(', ',',')
veges = veges.replace(' ,',',')


veges1 = veges.split(',')[0:7]
veges2 = veges.split(',')[7:]
```

```python
import re
n = 0
f = []
vegetable = []
for i in veges2:
    name = i    
    if name not in f:
        f.append(name)
        
        f10 = file23[file23['Food Name'].str.contains(name) | file23['Food Name'].str.contains(name.lower())]
        info = [name] + f10.loc[:,f10.columns[1:]].mean().tolist()
        vegetable.append(info)
```

```python
onion = file23[file23['Food Name'].str.contains('Onion') & file23['Food Name'].str.contains('raw')]
garlic = file23[file23['Food Name'].str.contains('Garlic')]
beetroot = file23[file23['Food Name'].str.contains('Beetroot') & file23['Food Name'].str.contains('raw')]

onion = ['Onion'] + onion.loc[:,onion.columns[1:]].mean().tolist()
garlic = ['Garlic'] + garlic.loc[:,garlic.columns[1:]].mean().tolist()
beetroot = ['Beetroot'] + beetroot.loc[:,beetroot.columns[1:]].mean().tolist()

vegetable = [onion] + [garlic] + [beetroot] + vegetable
```

```python
vegetable_df = pd.DataFrame(vegetable, columns=file23.columns).dropna().round(1).reset_index(drop=True)
vegetable_df.sort_values(by=['Total sugars'], ascending=False)
```

```python
keys = "Cereal, Grain, Breads, Biscuits, Cookies, Pasta, Nut,\
Almond meal, Amaranth flour, Barley including flour, Bran cereals, Bread, Granary bread, Multigrain bread\
, Naan, Oatmeal bread, Pumpernickel bread, Roti, Sourdough with kamut\
, Cashews, Chestnut flour, Cous cous, Einkorn flour, Freekeh, Gnocchi, Granola bar, Muesli cereal\
, Muesli bar, Pistachios, Rye, Rye crispbread, Semolina, Spelt flour"
keys = keys.replace(', ',',')
keys = keys.replace(' ,',',')
keys = list(set(keys.split(',')))
keys1 = []; keys2 = []

for i in keys:
    if " " in i:
        keys1.append(i)
    else:
        keys2 += [i] 
print(keys1, '\n', keys2)
```

```python
import re
n = 0
f = []
grain = []
for i in keys1:
    name = i    
    if name not in f:
        f.append(name)
        
        df = file23[file23['Food Name'].str.contains(name) | file23['Food Name'].str.contains(name.lower())]
        info = df.values.tolist()
        grain += info
```

```python
import re
n = 0
f = []
for i in keys2:
    name = i    
    if name not in f:
        f.append(name)
        
        df = file23[file23['Food Name'].str.contains(name) | file23['Food Name'].str.contains(name.lower())]
        info = df.values.tolist()
        grain += info
```

```python
grain_df = pd.DataFrame(grain, columns=file23.columns).dropna().drop_duplicates(inplace=False).round(1)
grain_df = grain_df.sort_values(by=['Total dietary fibre'], ascending=False).reset_index(drop=True)
grain_df.head(30)
```
